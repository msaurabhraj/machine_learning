# principle component Analysis


```python
#it reduces the dimension into components . components are the derivatives 
```

```python
from sklearn.decomposition import PCA
```

```python
df9 = pd.read_csv("Iris-f.csv")
df9.dropna(axis=0,inplace=True)
df9.shape
```

```python
x= df9.iloc[:,0:4]
pca = PCA(n_components=2) #these are component not the feature 
scaled =pca.fit_transform(x)
```

```python
df_new = pd.DataFrame(scaled)
df_new
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_new['Species']= le.fit_transform(df9[['Species']])
df_new
```

# preprocessed data
