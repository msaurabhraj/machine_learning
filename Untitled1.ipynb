# Find & List the most important 5 features to determine the Outcome variable


```python
from sklearn.feature_selection import SelectKBest,f_regression
x = df2.drop(['Outcome'],axis=1)
y = df2.Outcome
sk = SelectKBest(f_regression,k=5) #out of 5 variable select one best value
x_new = sk.fit_transform(x,y)
x_new
```

```python
new_df=pd.DataFrame(x_new)
new_df.head()
```

```python
sk.get_support(indices=True) #to koow the selected feature 
```

# Check with heatmap or correlation matrix to confirm the 5 selected features


```python
plt.figure(figsize=(12,6))
cor = new_df.corr()
sns.heatmap(cor,annot=True)
plt.show()
```

# Perform PCA and generate two components for the 5 features selected

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_model=pca.fit_transform(new_df)
pca_model
```

```python
df_new = pd.DataFrame(pca_model)
df_new
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_new['outcome']=le.fit_transform(df2[['Outcome']])
df_new
```

# split the data into train and test (75:25) keeping the target variable as Outcome


```python
x= df_new.iloc[:,0:2]
y= df_new['outcome']
from sklearn.model_selection  import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.75,shuffle=True,random_state=1)

```

```python
print(xtrain.shape) 
print(xtest.shape)
print(ytrain.shape) 
print(ytest.shape)
```

```python
xtrain.value_counts()
```

```python
from sklearn.linear_model import LinearRegression
slr= LinearRegression()
slr.fit(xtrain,ytrain)
```

```python
ypred= slr.predict(xtest)
from sklearn.metrics import r2_score
r2_score(ytest,ypred) #31% is the accuracy 
```
