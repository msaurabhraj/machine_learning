```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Data for Theorm Testing
# t1(n):23*n*n*n+43*n+20	g1(n)=n*n*n
# t2(n):7*n*n*n*n+4*n*n*n+2	g2(n)=n*n*n*n

limit = 60
```

```python
def t1(n):
    return 23*n*n*n + 43*n + 20

def g1(n):
    return n**3

def t2(n):
    return 7*n*n*n*n + 4*n*n*n + 2

def g2(n):
    return n**4

def T1PlusT2(n):
    return 7*(n**4) + 27*(n**3) + 43*n + 22
```

```python
fx1 = [t1(x) for x in range(limit)]
gx1 = [g1(x) for x in range(limit)]

fx2 = [t2(x) for x in range(limit)]
gx2 = [g2(x) for x in range(limit)]

# fx1(x) + fx2(x)
fx = [T1PlusT2(x) for x in range(limit)]

# max(gx1(x), gx2(x))
gx = gx2.copy()
```

```python
df = pd.DataFrame({"fx1(n)" : fx1, "fx2(n)" : fx2, "fx1(n)+fx2(n)" : fx, "max(gx1(n), gx2(n))" : gx2})
```

```python
df.plot.line()
plt.xlabel(">-------- n ------->")
plt.ylabel(">-------- f(n) ------->")
plt.show()
```
