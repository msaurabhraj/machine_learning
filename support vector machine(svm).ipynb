```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("fruits.csv")
df1.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df1["Class"]=le.fit_transform(df1['Class'])
df1.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
mn = MinMaxScaler()
df1['Weight'] = mn.fit_transform(df1[["Weight"]])
df1['Size'] = mn.fit_transform(df1[["Size"]])
df1.head()
```

```python
from sklearn.model_selection import train_test_split
x = df1[['Weight','Size']]
y = df1['Class']
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=0.25 ,shuffle=True ,random_state=1)
```

# Linear SVM

```python
from sklearn.svm import LinearSVC
clf = LinearSVC()
clf.fit(xtrain,ytrain)
```

```python
ypred = clf.predict(xtest)
ypred1 = clf.predict(xtrain)
from sklearn.metrics import accuracy_score
accuracy_score(ypred,ytest)
```

```python
print(clf.score(xtrain,ytrain))
print(clf.score(xtest,ytest))
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(np.array(xtrain),np.array(ytrain),clf)
```

# linear svm with svc class 

```python
from sklearn.svm import SVC # svm fro non linear 
clf1 = SVC(kernel="linear") #you can add all kernal trick here 
clf1.fit(xtrain,ytrain)
clf1.support_vectors_
```

```python
print(clf1.score(xtrain,ytrain))
print(clf1.score(xtest,ytest))
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(np.array(xtrain),np.array(ytrain),clf1)
```

# Kernal trick 

```python
clf2 =SVC(kernel="sigmoid").fit(xtrain,ytrain)
clf3 =SVC(kernel="poly").fit(xtrain,ytrain)
clf4 =SVC(kernel="rbf").fit(xtrain,ytrain)
```

```python
plot_decision_regions(np.array(xtrain),np.array(ytrain),clf2)
```

```python
plot_decision_regions(np.array(xtrain),np.array(ytrain),clf3)
```

```python
df5 = pd.read_csv("BMI.csv")
df5.head()
```
