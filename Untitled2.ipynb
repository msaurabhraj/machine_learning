```python
import pandas as pd
import seaborn as scs
```

```python
df = pd.read_csv('exp.csv')
```

```python
df
```

```python
x = df[['experience']]
y = df['salary']
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x,y)
print(lr.coef_,lr.intercept_)
```

```python
ypred= lr.predict(x)
from sklearn.metrics import r2_score,mean_absolute_error
print(r2_score(y,ypred))
print(mean_absolute_error(y,ypred))
```

```python
import matplotlib.pyplot as plt 
```

```python
scs.scatterplot(data=df,x='experience',y='salary')
plt.plot(x,ypred,color='r')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
pf=PolynomialFeatures(degree=5)
xp =pf.fit_transform(x)
pr = LinearRegression()
pr.fit(xp,y)
print(pr.coef_,pr.intercept_)
```

```python
p_ypred= pr.predict(xp)
from sklearn.metrics import r2_score,mean_absolute_error
print(r2_score(y,p_ypred))
print(mean_absolute_error(y,p_ypred))
```

```python
plt.scatter(x,y,color='r')
plt.plot(x,p_ypred,color='g')
plt.xlabel('Experience')
plt.ylabel("salary")
plt.plot()
```

```python
from sklearn.linear_model import Ridge
r = Ridge(alpha=0.5)
r.fit(xp,y)
print(r.coef_,r.intercept_)
```

```python
r_ypred= r.predict(xp)
from sklearn.metrics import r2_score,mean_absolute_error
print(r2_score(y,r_ypred))
print(mean_absolute_error(y,r_ypred))
```

```python
plt.scatter(x,y,color='r')
plt.plot(x,r_ypred,color='g')
plt.xlabel('Experience')
plt.ylabel("salary")
plt.plot()
```

```python
from sklearn.linear_model import ElasticNet
rl = ElasticNet(alpha=0.5)
rl.fit(xp,y)
print(rl.coef_,rl.intercept_)
```

```python
el_ypred= rl.predict(xp)
from sklearn.metrics import r2_score,mean_absolute_error
print(r2_score(y,el_ypred))
print(mean_absolute_error(y,el_ypred))
```

```python
plt.scatter(x,y,color='r')
plt.plot(x,el_ypred,color='g')
plt.xlabel('Experience')
plt.ylabel("salary")
plt.plot()
```
