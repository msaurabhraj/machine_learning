```python
import random
import sys
import time
import matplotlib.pyplot as plt

def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1

    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

def quickSort(arr, low, high):
    if low < high:
        pivot = partition(arr, low, high)
        quickSort(arr, low, pivot - 1)
        quickSort(arr, pivot + 1, high)

n_values = list(range(100, 10001, 500))
actual_times = []
estimated_times = []

for n in n_values:
    arr = random.sample(range(100, n * 1000, 100), n)
    start = time.time()
    quickSort(arr, 0, len(arr) - 1)
    end = time.time()

    actual_times.append(end - start)
    estimated_times.append(n * n)

# Plotting the graph
plt.plot(n_values, actual_times, label='Actual Time')
plt.plot(n_values, estimated_times, label='Estimated Time (O(n^2))')
plt.xlabel('Input Size (n)')
plt.ylabel('Time (seconds)')
plt.title('Quick Sort Time Complexity')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import random
import sys
import time
import matplotlib.pyplot as plt

def binarySearch(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1

n_values = list(range(100, 10001, 500))
actual_times = []
estimated_times = []

for n in n_values:
    arr = sorted(random.sample(range(n * 10), n))
    target = random.randint(0, n * 10)
    
    start = time.time()
    binarySearch(arr, target)
    end = time.time()

    actual_times.append(end - start)
    estimated_times.append(n)

# Plotting the graph
plt.plot(n_values, actual_times, label='Actual Time')
plt.plot(n_values, estimated_times, label='Estimated Time (O(log n))')
plt.xlabel('Input Size (n)')
plt.ylabel('Time (seconds)')
plt.title('Binary Search Time Complexity')
plt.legend()
plt.grid(True)
plt.show()
```
